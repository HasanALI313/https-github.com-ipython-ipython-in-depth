```python
vac=[180,172,178,185,190,195,192,200,210,190]
```

```python
print(vac)
```

```python
b=sum(vac)
```

```python
print(b)
```

```python
mean=b/len(vac)
```

```python
print(mean)
```

```python
var=sum((x-mean)**2 for x in vac)/len(vac)
```

```python
print(var)
```

```python
std=var**-0.5
```

```python
print(std)
```

```python
pos=mean+std
```

```python
print(pos)
```

```python
neg=(mean-std)
```

```python
print(neg)
```

```python

print(len(list(filter(lambda x: neg < x < pos, vac))))
```
